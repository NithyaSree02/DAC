```python
import pandas as pd
```

```python
data = pd.read_csv('statsfinal.csv')
```

```python
display(data)
```

```python
data.dropna(inplace=True)
```

```python
data.fillna(0, inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.to_csv('cleaned_data.csv', index=False)
```

```python
display(data)
```

```python
# Extract year from the 'Day' 'Month' 'year' from the 'Date' column using a lambda function
# We need to get the year from the data to analyse sales year to year
data['Day'] = data['Date'].apply(lambda x: x.split('-')[0])
data['Month'] = data['Date'].apply(lambda x: x.split('-')[1])
data['Year'] = data['Date'].apply(lambda x: x.split('-')[2])
data
```

```python
# Total unit sales Product 1, Product 2, Product 3, Product 4 
q = data[["Q-P1","Q-P2","Q-P3","Q-P4"]].sum()
print(q)
plt.figure(figsize=(8,8))
plt.pie(q,labels=data[["Q-P1","Q-P2","Q-P3","Q-P4"]].sum().index,shadow=True,autopct="%0.01f%%",textprops={"fontsize":20},wedgeprops={'width': 0.8},explode=[0,0,0,0.3])
plt.legend(loc='center right', bbox_to_anchor=(1.2, 0.8));
```

```python
# Total Revenue percent from sales from Product 1, Product 2, Product 3, Product 4  
s=data[["S-P1","S-P2","S-P3","S-P4"]].sum()
print(s)
plt.figure(figsize=(8,8))
plt.pie(s,labels=data[["S-P1","S-P2","S-P3","S-P4"]].sum().index,shadow=True,autopct="%0.01f%%",textprops={"fontsize":20},wedgeprops={'width': 0.8},explode=[0,0,0,0.3])
plt.legend(loc='center right', bbox_to_anchor=(1.2, 0.8))
```

```python
# which is the most occuring month 
print(data["Month"].value_counts())
plt.figure(figsize=(10,10))
sns.countplot(x="Month",data=data,edgecolor="black")
plt.xticks(rotation=90);
```

```python
# which is the most occuring Day 
print(data["Day"].value_counts())
plt.figure(figsize=(10,10))
sns.countplot(x="Day",data=data,edgecolor="black")
plt.xticks(rotation=90);
```

```python
# which is the most occuring year 
print(data["Year"].value_counts())
plt.figure(figsize=(10,10))
sns.countplot(x="Year",data=data,edgecolor="black")
plt.xticks(rotation=90);
```
